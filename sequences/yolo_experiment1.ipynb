```python
from ultralytics import YOLO

import cv2
import os
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import os
from PIL import Image

model = YOLO("yolo11n.pt")
poseModel = YOLO("yolo11n-pose.pt")
```

```python


sequence_path = "../soccernet_data/tracking/train/SNMOT-060"
img_dir = os.path.join(sequence_path, "img1")
img_files = [os.path.join(img_dir, f) for f in os.listdir(img_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
img_files = sorted(img_files)
```

```python
cv2.namedWindow("YOLO Tracking botsort", cv2.WINDOW_NORMAL | cv2.WINDOW_KEEPRATIO)
for img_path in img_files:
    img = cv2.imread(img_path)
    result = model.track(img, persist=True, show=False, tracker="botsort.yaml")[0]
    frame = result.plot()            # BGR
    cv2.imshow("YOLO Tracking", frame)
    key = cv2.waitKey(30) & 0xFF     # ~30 ms delay
    if key == ord("q"):
        break

cv2.destroyAllWindows()
```

```python
cv2.namedWindow("YOLO Tracking bytetrack", cv2.WINDOW_NORMAL | cv2.WINDOW_KEEPRATIO)
for img_path in img_files:
    img = cv2.imread(img_path)
    result = model.track(img, persist=True, show=False, tracker="bytetrack.yaml")[0]
    frame = result.plot()            # BGR
    cv2.imshow("YOLO Tracking", frame)
    key = cv2.waitKey(30) & 0xFF     # ~30 ms delay
    if key == ord("q"):
        break

cv2.destroyAllWindows()
```

```python
cv2.namedWindow("YOLO Tracking botsort", cv2.WINDOW_NORMAL | cv2.WINDOW_KEEPRATIO)
for img_path in img_files:
    img = cv2.imread(img_path)
    # result = model.track(img, persist=True, show=False, tracker="botsort.yaml")[0]
    # frame = result.plot() 
    # cv2.imshow("YOLO Tracking", frame)
    poseResult = poseModel.track(img, persist=True, show=False, tracker="botsort.yaml")[0]
    
    poseFrame = poseResult.plot()
    
    cv2.imshow("YOLO Pose Tracking", poseFrame)
    key = cv2.waitKey(30) & 0xFF     # ~30 ms delay
    if key == ord("q"):
        break

cv2.destroyAllWindows()
```
